### Option 3: Using Title

```python
df.head()
```

```python
def get_title(x):
    x1 = x.split(',')[1]
    x2 = x1.split('.')[0]
    return x2.strip().lower()
```

```python
df.Name.map(lambda x: get_title(x)).unique()
```

```python
def get_title2(x):
    title_group = {
        'mr'             : 'Mr',
        'mrs'            : 'Mrs',
        'miss'           : 'Miss',
        'master'         : 'Master',
        'don'            : 'Sir',
        'rev'            : 'Sir',
        'dr'             : 'Officer',
        'mme'            : 'Mrs',
        'ms'             : 'Mrs',
        'major'          : 'Officer',
        'lady'           : 'Lady',
        'sir'            : 'Sir',
        'mlle'           : 'Miss',
        'col'            : 'Officer',
        'capt'           : 'Officer',
        'the countess'   : 'Lady',
        'jonkheer'       : 'Sir',
        'dona'           : 'Lady'
    }
    x1 = x.split(',')[1]
    x2 = x1.split('.')[0]
    return title_group[x2.strip().lower()]
```

```python
df.Name.map(lambda x: get_title2(x)).unique()
```

```python
df['Title'] = df.Name.map(lambda x: get_title2(x))
```

```python
df.head()
```

```python
df[df.Age.notnull()].boxplot('Age','Title')
```

```python
title_age_median = df.groupby('Title').Age.transform('median')
```

```python
df.Age.fillna(title_age_median,inplace=True)
```

### Feature : IsMother

```python
df['IsMother'] = np.where((df.Sex == 'female') & (df.Parch > 0) & (df.Age > 18) & (df.Title != 'Miss'),1,0)
```

```python
pd.crosstab(df[df.Survived != -888].Survived,df[df.Survived != -888].IsMother)
```

### Feature : Dec

```python
df.head()
```

```python
df.Cabin.value_counts()
```

```python
def get_dec(cabin):
    if pd.notnull(cabin):
        return str(cabin).strip()[0].upper()
    else:
        return 'Z'
```

```python
df['Dec'] = df['Cabin'].map(lambda x: get_dec(x))
```

```python
df.Dec.value_counts()
```

```python
df.loc[df.Cabin == 'T','Dec'] = 'Z'
```

## Categorical Feature Encoding


```python
df['IsMale'] = np.where(df.Sex == 'male',1,0)
```

```python
df.head()
```

```python
df.rename(columns={'fare_bin':'Fare_Bin'},inplace=True)
```

```python
df = pd.get_dummies(df,columns=['Dec', 'Pclass', 'Title', 'Fare_Bin', 'Embarked', 'AgeState'])
```

```python
df.info()
```

```python
### Droping columns that are not required any more
df.drop(['Cabin','Name','Sex','Ticket','Parch','SibSp'],axis=1, inplace=True)
```

```python
columns =[column for column in df.columns if column!='Survived']
```

```python
columns = ['Survived'] + columns
df = df[columns]
```
